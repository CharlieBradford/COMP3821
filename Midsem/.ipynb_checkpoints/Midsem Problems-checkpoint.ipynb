```python
n = 1000
arr = [randint(0,1000000) for x in range(0, n*2)]
minarr = []
questions = 0
counter = n
maxf = max(arr)
minf = min(arr)
while len(arr) > 1:

    a = arr.pop()
    b = arr.pop()
    questions = questions + 1
    if a > b:
        bigger = a
        smaller = b
    else:
        bigger = b
        smaller = a
    arr.append(bigger)
    if counter > 0:
        minarr.append(smaller)
        counter = counter - 1
biggest = bigger
while len(minarr) > 1:
    a = minarr.pop()
    b = minarr.pop()
    questions = questions + 1
    if a > b:
        minarr.append(b)
    else:
        minarr.append(a)
smallest = minarr[0]

print("Questions asked: {}, max allowed {}".format(questions, (3*n)-2))
print("Max found {}, real max {}".format(biggest, maxf))
print("Min found {}, real min {}".format(smallest, minf))
```

```python
from random import shuffle
n = 10
questions = 0
arr = [x for x in range(0, 2**n)]
shuffle(arr)
maxf = 2**n - 1
max2f = 2**n - 2

defeated = {}

while len(arr) > 1:
    a = arr.pop()
    b = arr.pop()
    
    print("{}, {}".format(a, b))

    questions = questions + 1
    if a > b:
        bigger = a
        smaller = b
    else:
        bigger = b
        smaller = a
    if bigger not in defeated.keys():
        defeated[bigger] = []
    defeated[bigger].append(smaller)
    arr.append(bigger)

    


biggest = bigger

seclist = defeated[biggest]
print(len(seclist))

while len(seclist) > 1:
    a = seclist.pop()
    b = seclist.pop()
    questions = questions + 1
    if a > b:
        seclist.append(a)
    else:
        seclist.append(b)
second = seclist.pop()



print("Questions asked: {}, max allowed {}".format(questions, (2**n) + n - 2))
print("Max found {}, real max {}".format(biggest, maxf))
print("Sec found {}, real sec {}".format(second, max2f))
```
